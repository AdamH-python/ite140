```python
import pandas as pd
```

# Bringing in dat data

```python
stocks = pd.read_excel("Stocks Data.xlsx")
```

## Print a little bit of the table

```python
stocks
```

## Check the datatypes

```python
stocks.dtypes
```

## Setting Correct Datatypes

```python
stocks = stocks.astype({"symbol": "string"})
```

## Looking at some Summary Data

```python
stocks.loc[:,"close"].mean()
```

```python
stocks.loc[:, "high"].max()
```

```python
stocks.loc[:, "low"].min()
```

## Adding a Name to the Index

```python
stocks.index.name = "stock #"

stocks
```

## Adding a new column for the Total Value, also called the Market Cap

```python
stocks.loc[:, "total value"] = stocks.loc[:, "close"] * stocks.loc[:, "volume"]
stocks
```

## Looking at all the occurences of Apple Stocks, with the symbol AAPL

```python
apple_stock = stocks[stocks.symbol == "AAPL"]
apple_stock
```

## Looking at all the Unique Stock Symbols the First Time They Appear

```python
unique_stocks = stocks.drop_duplicates(["symbol"])
unique_stocks
```

## I Tried to Find the Last occurence of the Stock, but That Didn't Really Work

```python
unique_last = stocks.drop_duplicates(["symbol"]).duplicated(keep="last")
unique_last
```
